```python
# Cell 3

from PIL import Image
import pytesseract
from transformers import pipeline

# Loading image and extract text
image_path = "/content/drive/MyDrive/DataMorphAI /PDFs for v1/testTesseract.png"
image = Image.open(image_path)
extracted_text = pytesseract.image_to_string(image)

# Loading summarizer pipeline
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

if len(extracted_text) > 1000:
    extracted_text = extracted_text[:1000]

# Generate summary
summary = summarizer(extracted_text, max_length=150, min_length=30, do_sample=False)[0]['summary_text']

print("Extracted Text:\n", extracted_text)
print("\nSummary:\n", summary)
```
